```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
```

```python
# Load file
df = pd.read_excel("Analytic_test.xlsx", header=1, index_col=0)
df.head(10)
```

```python
df.info()
```

```python
df.describe()
```

```python
# The number of users registered in each of the companies
users = [df.active_Red.sum(),
         df.has_Green_id.sum(),
         df.has_Black_id.sum(),
         df.has_Blue_id.sum()]
print("\n Всего пользуются услугами Red:", users[0],
      "\n Всего пользуются услугами Green:",users[1],
      "\n Всего пользуются услугами Black:",users[2],
      "\n Всего пользуются услугами Blue:",users[3])
```

```python
# Total number of registered users
fig1, ax1 = plt.subplots()
fig1.suptitle('Количество зарегистрированных пользователей')
colors = ['r', 'g', 'k', 'b']
ax1.bar(['Red', 'Green', 'Black', 'Blue'], users, color=colors, alpha=0.7, width=0.4)
plt.show()
```

```python
# Melt the dataframe so that we can plot the 'is_msk' column as a categorical variable
melted_df = pd.melt(df, id_vars=['HID', 'is_msk'],
                    value_vars=['active_Red', 'has_Green_id',
                                'has_Black_id', 'has_Blue_id'],
                    var_name='company', value_name='has_account')

# Filter the dataframe to only include rows where 'has_account'
# is equal to 1, so that we only count users with registered accounts
filtered_df = melted_df[melted_df['has_account'] == 1]

# Use the countplot function in Seaborn to visualize the data
colors = ['r', 'g', 'k', 'b']
ax = sns.countplot(x='is_msk', hue='company', data=filtered_df, palette=colors, alpha=0.7)

# Set the x-axis tick labels
new_labels = ['Other', 'Moscow']
ax.set_xticklabels(new_labels)

# Change the legend to display only the company names
handles, labels = ax.get_legend_handles_labels()
new_labels = ['Red', 'Green', 'Black', 'Blue']
ax.legend(handles=handles, labels=new_labels)
ax.set_title('Количество зарегистрированных пользователей в Москве и за ее пределами')
plt.show()
```

```python
# 
gmv = [df.Green_gvm_total.sum(),
       df.Blue_gmv_total.sum(),
       df.Black_gvm_total.sum()]
gmv
```

```python
users = [df.has_Green_id.sum(),
         df.has_Blue_id.sum(),
         df.has_Black_id.sum()]
part = np.divide(gmv, users)
part
```

```python
fig1, ax1 = plt.subplots()
fig1.suptitle('Суммарный доход компаний')
colors = ['g', 'b', 'k']
ax1.bar(['Green', 'Blue', 'Black'], gmv, color=colors, alpha=0.7, width=0.6)

fig2, ax2 = plt.subplots()
fig2.suptitle('Средний доход компаний, нормированный на одного пользователя')
colors = ['g', 'b', 'k']
ax2.bar(['Green', 'Blue', 'Black'], part, color=colors, alpha=0.7, width=0.6)
plt.show()
```

```python
# Create a scatter plot of total gmv vs. number of successful orders
df['Green_gvm_order_count'] = pd.to_numeric(df['Green_gvm_order_count'], errors='coerce')
sns.scatterplot(x='Green_gvm_total', y='Green_gvm_order_count', data=df,  hue='is_msk')
plt.xlabel('Total GMV')
plt.ylabel('Successful Orders')

plt.title('Зависимость между общим gmv и количеством успешных заказов (Green)',size=14)

plt.show()
```

```python
# Create a scatter plot of total gmv vs. number of successful orders
df['Black_gvm_order_count'] = pd.to_numeric(df['Black_gvm_order_count'], errors='coerce')
sns.scatterplot(x='Black_gvm_total', y='Black_gvm_order_count', data=df,  hue='is_msk')
plt.xlabel('Total GMV')
plt.ylabel('Successful Orders')

plt.title('Зависимость между общим gmv и количеством успешных заказов (Black)',size=14)

plt.show()
```

```python
# Calculate the number of installations by platform
android_installs = df['Red_android'].sum()
apple_installs = df['Red_apple'].sum()
web_installs = df['Red_web'].sum()

# Create a pie chart showing the number of installations by platform
fig, ax = plt.subplots()
ax.pie([android_installs, apple_installs, web_installs],
       labels=['Android', 'Apple', 'Web'],
       autopct='%1.1f%%')
ax.set_title('Используемые платформы (Red)')

# Create a bar chart showing the distribution of users by location
fig, ax = plt.subplots()
ax.bar(['Moscow', 'Other'],
        [df[df['is_msk'] == 1].shape[0], df[df['is_msk'] == 0].shape[0]])
ax.set_title('Red Company Installations by Location')
ax.set_xlabel('Location')
ax.set_ylabel('Number of Installations')

plt.show()
```
